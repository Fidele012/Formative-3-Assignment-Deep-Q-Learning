# **Checking the Environment**

We test whether the Boxing environment loads correctly.
This confirms that the Atari game is ready for training.

```python
import gymnasium as gym
import ale_py
gym.register_envs(ale_py)

env = gym.make("ALE/Boxing-v5")
obs, info = env.reset()

print("Environment loaded successfully!")
print("Observation shape:", obs.shape)
print("Number of actions:", env.action_space.n)

env.close()
```

# **Creating the DQN Model**

We create a DQN agent that will learn how to play Boxing.
We use CnnPolicy because the game input is an image.

```python
from stable_baselines3 import DQN

env = gym.make("ALE/Boxing-v5")

model = DQN(
    "CnnPolicy",
    env,
    learning_rate=0.00005,
    buffer_size=10000,
    learning_starts=1000,
    batch_size=32,
    gamma=0.98,
    exploration_fraction=0.2,
    exploration_final_eps=0.05,
    verbose=1
)

print("Model created successfully!")
```

# **Training the Agent**

We now train the agent so it can learn from the Boxing environment.
At this stage, the model starts improving by trying actions and receiving rewards.

```python
model.learn(total_timesteps=10000)
```

# **Saving the Model**

We save the trained model so it can be used later for evaluation.
This model file will be loaded in the play script.

```python
model.save("dqn_boxing_model")
```

# **Loading the Model for Evaluation**

We load the saved model and prepare it for testing.
This allows the agent to play using what it has learned.

```python
from stable_baselines3 import DQN

loaded_model = DQN.load("dqn_boxing_model")
print("Model loaded successfully!")
```

# **Creating the Evaluation Environment**

We create the Boxing environment again for testing the trained agent.
This is the same environment used during training.

```python
eval_env = gym.make("ALE/Boxing-v5", render_mode="rgb_array")
obs, info = eval_env.reset()

print("Evaluation environment ready!")
```

# **Running the Agent**

We let the trained agent play Boxing using the learned policy.
The agent now chooses actions based on what it learned during training.

```python
obs, info = eval_env.reset()

for step in range(200):
    action, _states = loaded_model.predict(obs, deterministic=True)
    obs, reward, terminated, truncated, info = eval_env.step(action)

    if terminated or truncated:
        obs, info = eval_env.reset()

print("Evaluation run completed!")
eval_env.close()
```

# **Experiment 5**

We use a moderate gamma value to compare it with lower and higher reward focus.
This helps us understand how future reward weighting affects learning.

```python
import numpy as np

# create environment
env = gym.make("ALE/Boxing-v5")

# build model
model_exp5 = DQN(
    "CnnPolicy",
    env,
    learning_rate=0.0005,
    buffer_size=5000,
    learning_starts=1000,
    batch_size=32,
    gamma=0.97,
    exploration_fraction=0.2,
    exploration_final_eps=0.05,
    verbose=0
)

# train
model_exp5.learn(total_timesteps=10000)

# evaluate
eval_env = gym.make("ALE/Boxing-v5")

rewards = []

for episode in range(5):
    obs, info = eval_env.reset()
    done = False
    total_reward = 0

    while not done:
        action, _ = model_exp5.predict(obs, deterministic=True)
        obs, reward, terminated, truncated, info = eval_env.step(action)
        total_reward += reward
        done = terminated or truncated

    rewards.append(total_reward)

eval_env.close()

print("Final Average Reward:", np.mean(rewards))

model_exp5.save("exp5_boxing_model")

env.close()
```
